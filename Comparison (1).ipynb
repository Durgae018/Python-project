```python
def support_opposition(pos,neu,neg):
    p = []
    n = []
    num_elements = len(pos)
    num_values_per_sublist = min(25, len(pos[0]))  

    for j in range(num_elements):
        sub_p1 = []  
        sub_n1 = []
        for i in range(num_values_per_sublist):
            t = pos[j][i] + neu[j][i] + neg[j][i]
            a = pos[j][i] / t if t != 0 else 0
            b = neg[j][i] / t if t != 0 else 0
            sub_p1.append(a)
            sub_n1.append(b)
        p.append(sub_p1)  
        n.append(sub_n1)
    return p,n
```

```python
import math
import numpy as np
def formula(pos,neg):
    a=[]
    mp=np.mean(pos)
    mn=np.mean(neg)
    sp=np.std(pos)
    sn=np.std(neg)
    up=mp+(1.96*(sp/math.sqrt(10)))
    un=mp-(1.96*(sp/math.sqrt(10)))
    vp=mn+(1.96*(sn/math.sqrt(10)))
    vn=mn-(1.96*(sn/math.sqrt(10)))
    a=[[un,up],[vn,vp]]
    return a
```

```python
def distance_ideal(a,w):
    d_plus=0
    for i in range(len(a)):
        d_plus+= np.square(w[i]*(1-a[i]))
    return np.sqrt(d_plus)
    
    
def distance_antideal(a,w):
    d_minus=0
    for i in range(len(a)):
        d_minus+= np.square(w[i]*a[i])
    return np.sqrt(d_minus)
```

```python
#-----------------------------INTUITIONISTIC FUZZY TODIM---------------------
import numpy as np
cars_data={}
i=0
keys=['Honda','Hyundai','Isuzu','Kia','Land-rover','Nexon','Porsche','Swift','Tata','Thar','Volvo']
for k in keys:
    cars_data[k]=[p1[i],n1[i]]
    i=i+1
print(cars_data)

def TODIM(data):
    # Convert the data into numpy arrays for easier manipulation
    data = np.array(data)
    
    # Number of alternatives (cars)
    n_alternatives = data.shape[0]
    
    # Number of attributes
    n_attributes = data.shape[2]
    
    # Calculate the weights for each attribute
    weights = np.mean(data, axis=0)
    
    # Calculate the ideal and anti-ideal solutions for each attribute
    ideal_solution = np.max(data, axis=0)
    anti_ideal_solution = np.min(data, axis=0)
    
    # Calculate the separation measures for each alternative
    separation_measures = np.zeros(n_alternatives)
    for i in range(n_alternatives):
        separation_measures[i] = np.sum(
            np.sum(weights * (data[i] - anti_ideal_solution) /
                   (ideal_solution - anti_ideal_solution), axis=1)
        )
    return separation_measures

# Calculate TODIM scores for each car
car_scores = TODIM(list(cars_data.values()))


# Rank the cars based on their TODIM scores
ranked_cars = sorted(cars_data.keys(), key=lambda x: car_scores[list(cars_data.keys()).index(x)], reverse=True)

# Print the ranked cars
print("-----------CARS RANKED BASED ON INTUITIONISTIC FUZZY TODIM-----------:")
for i, car in enumerate(ranked_cars):
    print(f"{i+1}. {car}")
```
